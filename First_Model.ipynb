```python
import os
import json
import requests
from tqdm import tqdm
import zipfile
import io
import random
from PIL import Image
from collections import Counter


import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
# from PIL import Image
import numpy as np
from pycocotools.coco import COCO
# from tqdm import tqdm


import os
import matplotlib.pyplot as plt

from torch.utils.data import random_split

from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction



print("Number of GPU : ", torch.cuda.device_count())
print("Number of GPU : ", torch.cuda.get_device_name())

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print("Using device : ", device)



def download_subset_coco(num_images=5000):
    """
    Download a subset of COCO dataset
    """
    # Create directories
    os.makedirs('coco/images', exist_ok=True)
    os.makedirs('coco/annotations', exist_ok=True)

    # Download annotations file using requests
    annotation_url = "http://images.cocodataset.org/annotations/annotations_trainval2017.zip"
    print("Downloading annotation file...")

    try:
        response = requests.get(annotation_url, stream=True)
        if response.status_code == 200:
            # Write the zip file
            with open('annotations.zip', 'wb') as f:
                f.write(response.content)

            # Extract the zip file
            with zipfile.ZipFile('annotations.zip', 'r') as zip_ref:
                zip_ref.extractall('.')

            # Remove the zip file
            os.remove('annotations.zip')
        else:
            raise Exception(f"Failed to download annotations: Status code {response.status_code}")

        # Load annotations
        with open('annotations/captions_train2017.json', 'r') as f:
            annotations = json.load(f)

        # Get subset of images and their annotations
        subset_images = annotations['images'][:num_images]
        image_ids = [img['id'] for img in subset_images]
        subset_annotations = [ann for ann in annotations['annotations']
                            if ann['image_id'] in image_ids]

        # Create new annotation file with subset
        subset_data = {
            'info': annotations['info'],
            'licenses': annotations['licenses'],
            'images': subset_images,
            'annotations': subset_annotations
        }

        # Save subset annotations
        with open('coco/annotations/captions_subset_train2017.json', 'w') as f:
            json.dump(subset_data, f)

        # Download subset of images
        print(f"\nDownloading {num_images} images...")
        for img in tqdm(subset_images):
            img_url = f"http://images.cocodataset.org/train2017/{img['file_name']}"
            img_path = f"coco/images/{img['file_name']}"

            # Download image if it doesn't exist
            if not os.path.exists(img_path):
                try:
                    response = requests.get(img_url)
                    if response.status_code == 200:
                        with open(img_path, 'wb') as f:
                            f.write(response.content)
                except Exception as e:
                    print(f"Error downloading {img['file_name']}: {str(e)}")

        print("\nCleanup...")
        # Remove the annotations directory after we're done
        if os.path.exists('annotations'):
            import shutil
            shutil.rmtree('annotations')

        return 'coco/images', 'coco/annotations/captions_subset_train2017.json'

    except Exception as e:
        print(f"Error during download: {str(e)}")
        raise

# For testing the download function independently
if __name__ == "__main__":
    images_dir, annotations_file = download_subset_coco(num_images=6000) #Change here
    print(f"\nDownload complete!")
    print(f"Images directory: {images_dir}")
    print(f"Annotations file: {annotations_file}")





class ImageCaptioningDataset(Dataset):
    def __init__(self, root_dir, annotations_file, transform=None, max_samples=5000):
        self.root_dir = root_dir
        self.transform = transform

        print(f"Loading annotations from: {annotations_file}")
        self.coco = COCO(annotations_file)

        # Get image ids and limit to max_samples
        self.img_ids = list(self.coco.imgs.keys())[:max_samples]
        print(f"Number of images loaded: {len(self.img_ids)}")

        # Build vocabulary
        self.word_freq = Counter()
        self.captions = []
        self.annotations = {}
        print("Building vocabulary...")
        for img_id in tqdm(self.img_ids):
            ann_ids = self.coco.getAnnIds(imgIds=img_id)
            anns = self.coco.loadAnns(ann_ids)
            self.annotations[img_id] = anns  # Store all captions for each image
            for ann in anns:
                caption = ann['caption'].lower().split()
                self.word_freq.update(caption)
                self.captions.append(caption)

        # Create word to index mapping
        words = [word for word, freq in self.word_freq.items() if freq > 3]
        self.word2idx = {'<PAD>': 0, '<START>': 1, '<END>': 2, '<UNK>': 3}
        self.word2idx.update({word: idx + 4 for idx, word in enumerate(words)})
        self.idx2word = {v: k for k, v in self.word2idx.items()}
        print(f"Vocabulary size: {len(self.word2idx)}")

    def __len__(self):
        return len(self.img_ids)

    def __getitem__(self, idx):
        img_id = self.img_ids[idx]
        img_info = self.coco.loadImgs(img_id)[0]
        img_path = os.path.join(self.root_dir, img_info['file_name'])

        try:
            # Load and transform the image
            image = Image.open(img_path).convert('RGB')
            if self.transform:
                image = self.transform(image)

            # Randomly select one caption for the current image
            anns = self.annotations[img_id]
            selected_caption = random.choice(anns)['caption'].lower().split()

            # Convert the selected caption to indices
            caption = ['<START>'] + selected_caption + ['<END>']
            caption = [self.word2idx.get(word, self.word2idx['<UNK>']) for word in caption]

            return image, torch.tensor(caption)
        except Exception as e:
            print(f"Error loading image {img_path}: {str(e)}")
            return None

    def collate_fn(self, batch):
        batch = [item for item in batch if item is not None]
        if len(batch) == 0:
            return None

        images, captions = zip(*batch)
        images = torch.stack(images)

        caption_lengths = [len(cap) for cap in captions]
        max_length = max(caption_lengths)

        padded_captions = torch.zeros(len(captions), max_length).long()
        for i, cap in enumerate(captions):
            end = caption_lengths[i]
            padded_captions[i, :end] = cap[:end]

        return images, padded_captions




# Attention Mechanism
class Attention(nn.Module):
    def __init__(self, encoder_dim, decoder_dim, attention_dim):
        super(Attention, self).__init__()
        self.encoder_att = nn.Linear(encoder_dim, attention_dim)
        self.decoder_att = nn.Linear(decoder_dim, attention_dim)
        self.full_att = nn.Linear(attention_dim, 1)
        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, encoder_out, decoder_hidden):
            
        # Reshape encoder_out if necessary
        batch_size = encoder_out.size(0)
        encoder_dim = encoder_out.size(-1)
        num_pixels = encoder_out.size(1) * encoder_out.size(2)
        
        # Reshape encoder_out to (batch_size, num_pixels, encoder_dim)
        encoder_out = encoder_out.view(batch_size, -1, encoder_dim)
        
        # Ensure decoder_hidden has correct shape
        if decoder_hidden.dim() == 3:  # If (num_layers, batch_size, hidden_size)
            decoder_hidden = decoder_hidden[-1]  # Use last layer's hidden state
        
        # Transform encoder and decoder features to the same space
        att1 = self.encoder_att(encoder_out)  # (batch_size, num_pixels, attention_dim)
        att2 = self.decoder_att(decoder_hidden).unsqueeze(1)  # (batch_size, 1, attention_dim)
        
        # Calculate attention scores
        att = self.full_att(self.relu(att1 + att2)).squeeze(2)  # (batch_size, num_pixels)
        alpha = self.softmax(att)  # (batch_size, num_pixels)
        
        # Weight encoder outputs
        context = (encoder_out * alpha.unsqueeze(2)).sum(dim=1)  # (batch_size, encoder_dim)
        
        return context, alpha



# Encoder with ResNet
class EncoderCNN(nn.Module):
    def __init__(self, embed_size, encoder_dim=2048):
        super(EncoderCNN, self).__init__()
        self.encoder_dim = encoder_dim  # Store encoder_dim
        resnet = models.resnet50(pretrained=True)
        modules = list(resnet.children())[:-2]
        self.resnet = nn.Sequential(*modules)
        self.adaptive_pool = nn.AdaptiveAvgPool2d((14, 14))
        self.conv = nn.Conv2d(resnet.fc.in_features, self.encoder_dim, kernel_size=1)
        self.bn = nn.BatchNorm2d(self.encoder_dim)
        self.relu = nn.ReLU()

    def forward(self, images):
        with torch.no_grad():
            features = self.resnet(images)  # Shape: (batch_size, 2048, 7, 7)
        features = self.adaptive_pool(features)      # Shape: (batch_size, 2048, 14, 14)
        features = self.conv(features)               # Reduce dimension to encoder_dim
        features = self.bn(features)
        features = self.relu(features)
        features = features.permute(0, 2, 3, 1)      # Shape: (batch_size, 14, 14, encoder_dim)
        return features


# Decoder with Attention
class DecoderRNN(nn.Module):
    def __init__(self, embed_size, hidden_size, vocab_size, attention_dim, encoder_dim, num_layers=1):
        super(DecoderRNN, self).__init__()
        self.attention = Attention(encoder_dim, hidden_size, attention_dim)
        self.embedding = nn.Embedding(vocab_size, embed_size)
        self.lstm = nn.LSTM(embed_size + encoder_dim, hidden_size, num_layers, batch_first=True)
        self.linear = nn.Linear(hidden_size, vocab_size)
        self.dropout = nn.Dropout(0.5)

    def forward(self, encoder_out, captions):
        # Prepare embeddings
        embeddings = self.embedding(captions)  # (batch_size, max_caption_length, embed_size)

        # Initialize hidden states (h, c) to zeros
        batch_size = encoder_out.size(0)
        h, c = torch.zeros(self.lstm.num_layers, batch_size, self.lstm.hidden_size).to(encoder_out.device), \
               torch.zeros(self.lstm.num_layers, batch_size, self.lstm.hidden_size).to(encoder_out.device)

        # Output tensors
        outputs = []
        for t in range(embeddings.size(1)):  # Iterate over time steps
            context, _ = self.attention(encoder_out, h[-1])  # Use last hidden state
            lstm_input = torch.cat([embeddings[:, t, :], context], dim=1).unsqueeze(1)
            _, (h, c) = self.lstm(lstm_input, (h, c))
            outputs.append(self.linear(h[-1]))  # Use last layer's hidden state

        # Stack outputs and return
        return torch.stack(outputs, dim=1)  # (batch_size, max_caption_length, vocab_size)




class MetricsTracker:
    def __init__(self):
        self.train_epoch_losses = []
        self.test_epoch_losses = []
        self.train_batch_losses = []
        self.train_accuracies = []
        self.test_accuracies = []
        

    def update_batch_loss(self, loss, is_train=True):
        if is_train:
            self.train_batch_losses.append(loss)
        
    def update_epoch_metrics(self, epoch_loss, accuracy, is_train=True):
        if is_train:
            self.train_epoch_losses.append(epoch_loss)
            self.train_accuracies.append(accuracy)
        else:
            self.test_epoch_losses.append(epoch_loss)
            self.test_accuracies.append(accuracy)
        

    def plot_metrics(self, save_path=None):
        plt.figure(figsize=(15, 5))
        
        # Plot losses
        plt.subplot(1, 2, 1)
        plt.plot(self.train_epoch_losses, 'b-', label='Training Loss')
        plt.plot(self.test_epoch_losses, 'r-', label='Test Loss')
        plt.title('Loss Over Epochs')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)
        
        # Plot accuracies
        plt.subplot(1, 2, 2)
        plt.plot(self.train_accuracies, 'g-', label='Training Accuracy')
        plt.plot(self.test_accuracies, 'y-', label='Test Accuracy')
        plt.title('Accuracy Over Epochs')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy (%)')
        plt.legend()
        plt.grid(True)
        
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path)
        plt.show()


# -----------------------------------------------------------------------------------------------------------------------------------------


def generate_caption(encoder, decoder, image_input, transform, word2idx, idx2word, device, max_length=50):
    encoder.eval()
    decoder.eval()

    # Handle tensor input directly
    if isinstance(image_input, torch.Tensor):
        image = image_input.to(device)
    else:  # Handle image path input
        image = Image.open(image_input).convert('RGB')
        image = transform(image).unsqueeze(0).to(device)

    # Pass through encoder
    with torch.no_grad():
        encoder_out = encoder(image)  # Shape: (batch_size, 14, 14, encoder_dim)
        encoder_out = encoder_out.view(encoder_out.size(0), -1, encoder_out.size(-1))  # Flatten spatial dimensions
        batch_size = encoder_out.size(0)

        # Initialize the caption with START token
        caption = [word2idx['<START>']]

        # Initialize hidden states (h, c) for LSTM
        h = torch.zeros(decoder.lstm.num_layers, batch_size, decoder.lstm.hidden_size).to(device)
        c = torch.zeros(decoder.lstm.num_layers, batch_size, decoder.lstm.hidden_size).to(device)

        # Generate words until END token or max length
        for _ in range(max_length):
            embeddings = decoder.embedding(torch.tensor([caption[-1]]).to(device))  # Get embedding for last word
            context, _ = decoder.attention(encoder_out, h[-1])  # Compute attention context
            lstm_input = torch.cat([embeddings, context], dim=1).unsqueeze(1)  # Combine embeddings and context
            _, (h, c) = decoder.lstm(lstm_input, (h, c))  # Update LSTM states
            output = decoder.linear(h[-1])  # Project hidden state to vocabulary size
            predicted = output.argmax(dim=1).item()  # Get predicted word index
            caption.append(predicted)
            if predicted == word2idx['<END>']:  # Stop if END token is predicted
                break

    # Convert indices to words
    caption_words = [idx2word[idx] for idx in caption[1:-1]]  # Exclude START and END tokens
    return ' '.join(caption_words)









# BLEU Score Calculation with Smoothing
def calculate_bleu(predicted_caption, actual_captions, max_order=4):
    reference = [cap.split() for cap in actual_captions]
    candidate = predicted_caption.split()
    smoothing = SmoothingFunction().method4  # Apply smoothing to handle zero n-gram overlaps
    weights = [1.0 / max_order] * max_order  # Equal weight for 1-gram to max_order-gram
    return sentence_bleu(reference, candidate, weights=weights, smoothing_function=smoothing)



# Evaluate BLEU Score on Test Set
def evaluate_bleu(test_loader, encoder, decoder, dataset, transform, device):
    encoder.eval()
    decoder.eval()
    bleu_scores = []

    with torch.no_grad():
        for images, captions in test_loader:
            images = images.to(device)
            for idx in range(images.size(0)):
                image = images[idx].unsqueeze(0)
                actual_captions = [
                    ' '.join([dataset.idx2word[token.item()] for token in cap if token != dataset.word2idx['<PAD>']])
                    for cap in captions[idx].unsqueeze(0)
                ]

                predicted_caption = generate_caption(
                    encoder, decoder, image, transform, dataset.word2idx, dataset.idx2word, device
                )
                bleu = calculate_bleu(predicted_caption, actual_captions)
                bleu_scores.append(bleu)

    avg_bleu = sum(bleu_scores) / len(bleu_scores)
    print(f"Average BLEU Score: {avg_bleu:.4f}")
    return avg_bleu


def calculate_accuracy(outputs, targets, pad_value=0):
    """Calculate accuracy excluding padding tokens"""
    # Get predictions
    _, predicted = outputs.max(1)
    
    # Create mask for non-padding tokens
    mask = (targets != pad_value)
    
    # Calculate accuracy only on non-padding tokens
    correct = ((predicted == targets) * mask).sum().item()
    total = mask.sum().item()
    
    return (correct / total * 100) if total > 0 else 0


# -----------------------------------------------------------------------------------------------------------------------------------------


# Modified training function to include validation
def train_model(encoder, decoder, train_loader, test_loader, criterion, 
                encoder_optimizer, decoder_optimizer, device, num_epochs, patience, clip_value):
    
    print("Starting training...")
    metrics_tracker = MetricsTracker()
    best_test_loss = float('inf')
    early_stopping_counter = 0

    # Learning rate schedulers
    encoder_scheduler = torch.optim.lr_scheduler.StepLR(encoder_optimizer, step_size=10, gamma=0.5)
    decoder_scheduler = torch.optim.lr_scheduler.StepLR(decoder_optimizer, step_size=10, gamma=0.5)

    for epoch in range(num_epochs):
        # Training phase
        encoder.train()
        decoder.train()
        total_train_loss = 0
        total_train_accuracy = 0
        n_train_batches = 0
        errors = []  # Log errors for debugging

        progress_bar = tqdm(train_loader, 
                            desc=f'Epoch {epoch+1}/{num_epochs} - Training',
                            position=0, 
                            leave=True,
                            ncols=100)

        for i, batch in enumerate(progress_bar):
            if batch is None:
                continue

            images, captions = batch
            images = images.to(device)
            captions = captions.to(device)

            encoder_optimizer.zero_grad()
            decoder_optimizer.zero_grad()

            try:
                # Forward pass
                features = encoder(images)  # Shape: (batch_size, 14, 14, encoder_dim)
                features = features.view(features.size(0), -1, features.size(-1))  # (batch_size, num_pixels, encoder_dim)
                
                inputs = captions[:, :-1]  # Remove the last token for inputs
                targets = captions[:, 1:]  # Remove the first token for targets
                
                outputs = decoder(features, inputs)  # Ensure this matches the shape expected
                outputs_flat = outputs.reshape(-1, outputs.size(2))  # Flatten for CrossEntropyLoss
                targets_flat = targets.reshape(-1)

                # Loss and Backdrop
                loss = criterion(outputs_flat, targets_flat)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(encoder.parameters(), clip_value)
                torch.nn.utils.clip_grad_norm_(decoder.parameters(), clip_value)
                encoder_optimizer.step()
                decoder_optimizer.step()

                # Accuracy
                batch_accuracy = calculate_accuracy(outputs_flat, targets_flat)

                # Track metrics
                loss_value = loss.item()
                if not torch.isinf(torch.tensor(loss_value)) and not torch.isnan(torch.tensor(loss_value)):
                    total_train_loss += loss_value
                    total_train_accuracy += batch_accuracy
                    n_train_batches += 1
                    metrics_tracker.update_batch_loss(loss_value, is_train=True)

                progress_bar.set_postfix({
                    'Train Loss': f'{loss_value:.4f}',
                    'Train Acc': f'{batch_accuracy:.2f}%'
                })

            except Exception as e:
                errors.append(f"Batch {i}: {str(e)}")  # Log errors
                continue

        # Print logged errors at the end of the epoch
        if errors:
            print(f"Epoch {epoch+1} encountered {len(errors)} errors:")
            for error in errors:
                print(error)

        # Testing phase
        encoder.eval()
        decoder.eval()
        total_test_loss = 0
        total_test_accuracy = 0
        bleu_scores = []
        n_test_batches = 0

        with torch.no_grad():
            progress_bar = tqdm(test_loader,
                                desc=f'Epoch {epoch+1}/{num_epochs} - Testing',
                                position=0,
                                leave=True,
                                ncols=100)

            for i, batch in enumerate(progress_bar):
                if batch is None:
                    continue

                images, captions = batch
                images = images.to(device)
                captions = captions.to(device)

                try:
                    # Forward pass
                    features = encoder(images)
                    features = features.view(features.size(0), -1, features.size(-1))  # Reshape as above
                    
                    inputs = captions[:, :-1]
                    targets = captions[:, 1:]
                    outputs = decoder(features, inputs)

                    outputs_flat = outputs.reshape(-1, outputs.size(2))
                    targets_flat = targets.reshape(-1)

                    # Calculate loss and accuracy
                    loss = criterion(outputs_flat, targets_flat)
                    batch_accuracy = calculate_accuracy(outputs_flat, targets_flat)

                    # Track metrics
                    loss_value = loss.item()
                    if not torch.isinf(torch.tensor(loss_value)) and not torch.isnan(torch.tensor(loss_value)):
                        total_test_loss += loss_value
                        total_test_accuracy += batch_accuracy
                        n_test_batches += 1

                    # Generate BLEU score
                    for idx in range(images.size(0)):
                        image = images[idx].unsqueeze(0)
                        actual_captions = [
                            ' '.join([dataset.idx2word[token.item()] for token in cap if token != dataset.word2idx['<PAD>']])
                            for cap in captions[idx].unsqueeze(0)
                        ]
                        predicted_caption = generate_caption(
                            encoder, decoder, image, transform, dataset.word2idx, dataset.idx2word, device
                        )
                        bleu = calculate_bleu(predicted_caption, actual_captions)
                        bleu_scores.append(bleu)

                    progress_bar.set_postfix({
                        'Test Loss': f'{loss_value:.4f}',
                        'Test Acc': f'{batch_accuracy:.2f}%'
                    })

                except Exception as e:
                    errors.append(f"Test Batch {i}: {str(e)}")
                    continue

        # Update schedulers
        encoder_scheduler.step()
        decoder_scheduler.step()

        # Calculate and store metrics
        avg_train_loss = total_train_loss / n_train_batches if n_train_batches > 0 else float('inf')
        avg_train_accuracy = total_train_accuracy / n_train_batches if n_train_batches > 0 else 0
        avg_test_loss = total_test_loss / n_test_batches if n_test_batches > 0 else float('inf')
        avg_test_accuracy = total_test_accuracy / n_test_batches if n_test_batches > 0 else 0
        avg_bleu_score = sum(bleu_scores) / len(bleu_scores) if bleu_scores else 0

        metrics_tracker.update_epoch_metrics(avg_train_loss, avg_train_accuracy, is_train=True)
        metrics_tracker.update_epoch_metrics(avg_test_loss, avg_test_accuracy, is_train=False)

        print(f'\nEpoch [{epoch+1}/{num_epochs}]')
        print(f'Train Loss: {avg_train_loss:.4f}, Train Accuracy: {avg_train_accuracy:.2f}%')
        print(f'Test Loss: {avg_test_loss:.4f}, Test Accuracy: {avg_test_accuracy:.2f}%')
        print(f'BLEU Score: {avg_bleu_score:.4f}')

        # Save final checkpoint
        if epoch == num_epochs - 1:
            checkpoint_path = f'final_checkpoint.pth'
            torch.save({
                'epoch': num_epochs,
                'encoder_state_dict': encoder.state_dict(),
                'decoder_state_dict': decoder.state_dict(),
                'encoder_optimizer_state_dict': encoder_optimizer.state_dict(),
                'decoder_optimizer_state_dict': decoder_optimizer.state_dict(),
                'train_loss': avg_train_loss,
                'test_loss': avg_test_loss,
                'train_accuracy': avg_train_accuracy,
                'test_accuracy': avg_test_accuracy,
                'avg_bleu_score': avg_bleu_score,
                'embed_size': embed_size,
                'hidden_size': hidden_size,
                'attention_dim': attention_dim,
                'encoder_dim': encoder_dim,
                'vocab_size': len(dataset.word2idx),
                'word2idx': dataset.word2idx,
                'idx2word': dataset.idx2word,
                'num_layers': num_layers,
            }, checkpoint_path)

            print(f'Final model checkpoint saved: {checkpoint_path}')


    # Plot final metrics
    metrics_tracker.plot_metrics(save_path=f'metrics_epoch_{num_epochs}.png')

    return metrics_tracker



device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Hyperparameters
embed_size = 256
hidden_size = 512
num_layers = 2
learning_rate = 0.001
num_epochs = 150 #10
batch_size = 64 #16
attention_dim = 512
encoder_dim = 512 #2048
patience = 10
clip_value = 5.0
num_samples = 6000 #5000 # This loads this many images...
# num_images = 1000# This downloads this many images... go up and change where the function is called. Won't work from here

# Image transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),  # Data augmentation
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # Data augmentation
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Create dataset and dataloader
dataset = ImageCaptioningDataset(
    root_dir='coco/images',
    annotations_file='coco/annotations/captions_subset_train2017.json',
    transform=transform,
    max_samples=num_samples
)



# Split dataset into train and test sets
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])



train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    collate_fn=dataset.collate_fn,
    num_workers=4  # Set to 0 for debugging
)

test_loader = DataLoader(
    test_dataset,
    batch_size=batch_size,
    shuffle=False,  
    collate_fn=dataset.collate_fn,
    num_workers=4
)



# Initialize models
encoder = EncoderCNN(embed_size, encoder_dim=encoder_dim).to(device)
decoder = DecoderRNN(embed_size, hidden_size, len(dataset.word2idx), attention_dim, encoder_dim, num_layers).to(device)

# Loss and optimizers
criterion = nn.CrossEntropyLoss(ignore_index=dataset.word2idx['<PAD>'])
encoder_optimizer = torch.optim.AdamW(encoder.parameters(), lr=learning_rate, weight_decay=0.01)
decoder_optimizer = torch.optim.AdamW(decoder.parameters(), lr=learning_rate, weight_decay=0.01)



# Train the model
# train_model(encoder, decoder, train_loader, criterion, encoder_optimizer,
#             decoder_optimizer, device, num_epochs)

# Train the model with both train and test loaders
metrics = train_model(encoder, decoder, train_loader, test_loader, criterion, 
                     encoder_optimizer, decoder_optimizer, device, num_epochs, patience, clip_value)

# Evaluate BLEU score on test set
bleu_score = evaluate_bleu(test_loader, encoder, decoder, dataset, transform, device)
print(f"Final BLEU Score on Test Set: {bleu_score:.4f}")





def visualize_attention(encoder, decoder, image_path, transform, word2idx, idx2word, device):
    encoder.eval()
    decoder.eval()
    
    # Load and transform image
    image = Image.open(image_path).convert('RGB')
    transformed_image = transform(image).unsqueeze(0).to(device)
    
    with torch.no_grad():
        # Get encoder output
        encoder_out = encoder(transformed_image)
        
        # Initialize LSTM states
        batch_size = encoder_out.size(0)
        h = torch.zeros(num_layers, batch_size, decoder.lstm.hidden_size).to(device)
        c = torch.zeros(num_layers, batch_size, decoder.lstm.hidden_size).to(device)
        
        # Start with START token
        caption = [word2idx['<START>']]
        attention_weights = []
        
        for _ in range(50):  # Max caption length
            # Get current token embedding
            token = torch.tensor([caption[-1]]).to(device)
            token_embedding = decoder.embedding(token)
            
            # Get attention context
            context, alpha = decoder.attention(encoder_out, h[-1])  # Use last hidden state
            
            # Prepare LSTM input
            lstm_input = torch.cat([token_embedding, context], dim=1).unsqueeze(1)
            
            # LSTM forward pass
            output, (h, c) = decoder.lstm(lstm_input, (h, c))
            
            # Get prediction
            predictions = decoder.linear(output.squeeze(1))
            predicted = predictions.argmax(dim=1).item()
            
            # Store results
            caption.append(predicted)
            attention_weights.append(alpha.cpu().numpy())
            
            if predicted == word2idx['<END>']:
                break
    
    # Convert indices to words (excluding START and END)
    caption_words = [idx2word[idx] for idx in caption[1:-1]]
    
    # Visualize
    plt.figure(figsize=(12, 8))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    attention_map = np.mean(np.array(attention_weights), axis=0)
    attention_map = attention_map.reshape(14, 14)  # Reshape to match feature map size
    plt.imshow(image)
    plt.imshow(attention_map, cmap='hot', alpha=0.6)
    plt.title('Average Attention Map')
    plt.axis('off')
    
    plt.suptitle(' '.join(caption_words), y=1.05)
    plt.tight_layout()
    plt.show()
    
    return caption_words, attention_weights





def test_model(checkpoint_path, test_image_path, visualize=False):
    # Load the checkpoint
    checkpoint = torch.load(checkpoint_path)

    # Extract model parameters from the checkpoint
    embed_size = checkpoint.get('embed_size', 256)
    hidden_size = checkpoint.get('hidden_size', 512)
    attention_dim = checkpoint.get('attention_dim', 256)  # Default to 256 if not saved
    encoder_dim = checkpoint.get('encoder_dim', 512)
    vocab_size = checkpoint.get('vocab_size', 10000)
    num_layers = checkpoint.get('num_layers', 1)  # Default to 1 if not saved

    
    # print('num_layers from test_model func ----->', num_layers)
    
    # Initialize models with the same parameters as training
    encoder = EncoderCNN(embed_size, encoder_dim).to(device)
    decoder = DecoderRNN(embed_size, hidden_size, vocab_size, attention_dim, encoder_dim, num_layers).to(device)

    # Load the saved state
    encoder.load_state_dict(checkpoint['encoder_state_dict'])
    decoder.load_state_dict(checkpoint['decoder_state_dict'])

    # Generate caption and optionally visualize attention
    if visualize:
        caption_words, attention_weights = visualize_attention(
            encoder, decoder, test_image_path, transform, dataset.word2idx, dataset.idx2word, device
        )
        print(f"Generated caption with attention visualization: {' '.join(caption_words)}")
    else:
        caption = generate_caption(encoder, decoder, test_image_path, transform,
                                   dataset.word2idx, dataset.idx2word, device)
        print(f"Generated caption: {caption}")




def test_random_image(visualize=False):
    # Get a random image from the dataset
    random_idx = random.randint(0, len(dataset.img_ids) - 1)
    img_id = dataset.img_ids[random_idx]
    img_info = dataset.coco.loadImgs(img_id)[0]
    img_path = os.path.join(dataset.root_dir, img_info['file_name'])

    # Get actual captions for comparison
    ann_ids = dataset.coco.getAnnIds(imgIds=img_id)
    anns = dataset.coco.loadAnns(ann_ids)
    actual_captions = [ann['caption'] for ann in anns]

    # Test the model with or without attention visualization
    checkpoint_filename = f"final_checkpoint.pth"
    test_model(checkpoint_filename, img_path, visualize=visualize)

    # Print actual captions for comparison
    print("\nActual captions:")
    for i, caption in enumerate(actual_captions, 1):
        print(f"{i}. {caption}")

    # # Display the image
    # img = Image.open(img_path)
    # display(img)






# Run the test
test_random_image(visualize=True)
```
